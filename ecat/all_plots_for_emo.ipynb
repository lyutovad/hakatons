```python
import numpy as np
import pandas as pd
import plotly
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

import random
```

```python
df = pd.read_csv('12345.csv')
```

```python
df_group = df.groupby("person_id").mean()
```

```python
df_group_t = df_group.T
```

```python
emotions_list = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
```

```python
def id_to_emostate(id_num):
    fig = go.Figure()
    pull = [0]*7
    pull[np.argmax(df_group.T.drop('frame_num', axis=0)[id_num])] = 0.1
    fig.add_trace(go.Pie(values=df_group.T.drop('frame_num', axis=0)[id_num], labels=emotions_list, title=f'Эмоциональное состояние ученика id={id_num}', pull=pull))
    fig.update_traces(textposition='inside', textinfo='percent+label')
    fig.update_layout(font=dict(size=28))
    fig.show()
```

```python
def id_to_timedata(id_num): 
    df_pers =  df.loc[df['person_id']==id_num]
    fig = go.Figure()
    for emotion in emotions_list:
        fig.add_trace(go.Scatter(x=df_pers['frame_num'], y=df_pers[emotion], mode='markers+lines', name=emotion))
    fig.show()
        
```

```python
def id_to_timestate(id_num):
    df_pers =  df.loc[df['person_id']==id_num]
    fig = px.scatter(df_pers, y="dominant_emotion", x="frame_num", color="dominant_emotion", symbol="dominant_emotion")
    fig.update_traces(marker_size=12)
    fig.show()        
```

```python
def all_about(id_num):
    id_to_timedata(id_num)
    id_to_timestate(id_num)
    id_to_emostate(id_num)
```

```python
all_about(0)
```

```python
def mean_emotions(df):
    df[emotions_list].mean()
    fig = go.Figure()
    pull = [0]*7
    pull[np.argmax(df[emotions_list].mean().tolist())] = 0.1
    fig.add_trace(go.Pie(values=df[emotions_list].mean().tolist(), labels=emotions_list, title=f'Общее эмоциональное состояние', pull=pull))
    fig.update_traces(textposition='inside', textinfo='percent+label')
    fig.update_layout(font=dict(size=28))
    fig.show()
```

```python
mean_emotions(df)
```
